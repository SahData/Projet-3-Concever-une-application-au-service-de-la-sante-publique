### <span style='color:blue '>Test ANOVA (Analyse de la variance) </span>

Nous allons vérifier si nos variables nous permettent d’appliquer un test ANOVA pour voir l’effet de notre variable "nutriscore_grade" sur les variables numériques.
En premier lieu, pour appliquer ce test, la distribution des variables doit être normale. 
Examinons ceci :


#### Les variables quantitatives ont-elles une distribution normale ?

#### Test de normalité de Pearson

```python
numeric_columns = ['additives_n','nutriscore_score', 'energy_100g',
       'fat_100g','saturated-fat_100g','cholesterol_100g',
       'carbohydrates_100g','sugars_100g','proteins_100g','salt_100g',
       'sodium_100g','calcium_100g']
```

```python
from scipy import stats

numeric_columns = numeric_columns[1:]
for column in numeric_columns:
    print('________\n{}'.format(column))
    k2, p = stats.normaltest(df_fromages_clean[column], 
                             axis=0,
                             nan_policy = 'omit')
    alpha = 5e-2
    print("p = {:g}".format(p))
    if p < alpha:  # null hypothese: x a une distribution normale
        print("H0 est rejetée : {} n'est pas de distribution normale".format(column))
    else:
        print("H0 ne peut être rejetée :{}, on considère l'hypothèse de normalité".format(column))
```

- Les données ne sont pas issus d'une distribution normale

#### Test de homogénéité de Levene

On va varifier si les variables ont une homogénéité de variance; les variances de population sont égales. Pour tester cela, on utilise le test de Levene avec scipy:

```python
from scipy.stats import levene
a = df_fromages_clean['fat_100g']
b = df_fromages_clean['energy_100g']
c = df_fromages_clean['salt_100g']
d = df_fromages_clean['sodium_100g']
e = df_fromages_clean['saturated-fat_100g']
f = df_fromages_clean['nutriscore_score']
g = df_fromages_clean['cholesterol_100g']

stat, p = levene(a, b, c, d, e, f, g)
p
```

- La faible valeur p suggère que les variables n'ont pas des variances égales. 

Ce n'est pas surprenant en considérant la variance des variables :

```python
[np.var(x, ddof=1) for x in [a, b, c, d, e, f, g]]
```

- Il faut fournir en entrée au test les données pour chaque groupe

```python
from scipy import stats

data_d = df_fromages_clean[df_fromages_clean['nutriscore_grade'] == 'd'].select_dtypes(['float64']).dropna(axis=0).dropna(axis=1)
data_c = df_fromages_clean[df_fromages_clean['nutriscore_grade'] == 'c'].select_dtypes(['float64']).dropna(axis=0).dropna(axis=1)
data_e = df_fromages_clean[df_fromages_clean['nutriscore_grade'] == 'e'].select_dtypes(['float64']).dropna(axis=0).dropna(axis=1)


f_statistic, p = stats.f_oneway(
                                    data_d,
                                    data_c,
                                    data_e)
```

```python
pd.DataFrame({'features': data_d.columns.tolist(),
              'p' : p,
              'bool_test' : p<alpha})  #si p<alpha : on rejette l'hypothèse d'indépendance H0 : les variables ne sont pas indépendantes
```

- En l'absence de validation de l'hypothèse de normalité et d'homogénéité, ce test n'est pas applicable mais si c'était le cas, on pourrait interpréter les résultats comme suit : toutes les variables à l'exception de "additives_n" ont une distribution qui change en fonction des valeurs de "nutriscore_grade" (groupes de nutrition)

### <span style='color:blue '> Analyse en composantes principales (ACP)</span>

Pour vérifier s’il y a des liaisons entre nos variables, de façon que l’on puisse les regrouper en nouvelles variables synthétique, aussi réduire la dimension de notre data frame on procède à une ACP.

```python
import sklearn
from sklearn import preprocessing, decomposition
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from matplotlib.collections import LineCollection
from scipy.cluster.hierarchy import dendrogram
```

```python
df_fromages_clean_copy=df_fromages_clean.copy()  
```

```python
numeric_columns
```

```python
# selection des colonnes à prendre en compte dans l'ACP
data_pca = df_fromages_clean[['nutriscore_score','energy_100g','fat_100g','saturated-fat_100g','carbohydrates_100g',
 'sugars_100g','cholesterol_100g','proteins_100g','salt_100g','sodium_100g','calcium_100g']]
```

```python
colonnes_pca = data_pca.columns
```

```python
data_pca.select_dtypes(['float64']).shape
```

```python
# Centrage et Réduction
scaler = StandardScaler()
data_pca = scaler.fit_transform(data_pca)
```

```python
# Calcul des composantes principales
pca = PCA(n_components = data_pca.shape[1])
pca.fit(data_pca)
```

```python
pca.explained_variance_ratio_ * 100
```

```python
pca.components_.shape
```

```python
pca.components_
```

On va représenter graphiquement ces proportions de variances expliquées 

```python
plt.figure(figsize=(12,5))
plt.title('Somme cumulée de variance des composantes de l\'ACP')
plt.xlabel('Nombre de composantes de l\'ACP')
plt.ylabel('Variance cumulée')
plt.plot(np.linspace(1, 11,11),
         np.cumsum(pca.explained_variance_ratio_))
```

À partir de 7 variables, on a une variance cumulée de plus de 95 %. On pourrait donc réduire notre jeu de données à 7 dimensions si on le souhaitait.

```python
# Cercle des corrélations
from matplotlib.patches import Circle
sns.set_context("paper", font_scale=1.5) 
plt.figure(figsize=(15,15))
chaine = "Plan factoriel pour les 2 premières composantes de l'ACP"
plt.title(chaine)
plt.xlabel('Composante 1')
plt.ylabel('Composante 2')
ax = plt.gca()
ax.add_patch(Circle([0,0], radius=1, color='blue', fill=False, linestyle='-', clip_on=False))
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
x_pca = pca.components_[0,:]
y_pca = pca.components_[1,:]

sns.scatterplot(x = x_pca,
                y = y_pca,
               alpha = 0.7)
plt.plot([-1,1],[0,0], color='grey', alpha=0.8)
plt.plot([0,0],[-1,1], color='grey', alpha=0.8)

i = 0
for x,y,colonne in zip(x_pca,y_pca, colonnes_pca):

    plt.annotate(colonne, 
                 (x,y), 
                 textcoords="offset points", 
                 xytext=(0,5+np.random.randint(-10,10)),
                 ha='center') 
    ax.arrow(0, 0, x, y, head_width=0, head_length=0, fc='grey', ec='grey',alpha=0.2)


sns.set_context("paper", font_scale=1)
```
